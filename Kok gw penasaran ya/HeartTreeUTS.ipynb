```python
import pandas as pd #utk mengelola data set
import numpy as np #utk menampilkan array
import matplotlib.pyplot as plt #gambar plot, grafik, chart, tree
```

```python
df = pd.read_csv("heart_2020_cleaned.csv")
```

```python
col = ['HeartDisease', 'BMI', 'Smoking', 'AlcoholDrinking', 'Stroke', 'PhysicalHealth', 'MentalHealth'
      , 'DiffWalking', 'Sex', 'AgeCategory', 'Race', 'Diabetic', 'PhysicalActivity', 'GenHealth', 'SleepTime'
      , 'Asthma', 'KidneyDisease', 'SkinCancer']

for i in col :
  print(i, " : ", df[i].unique())
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

categorical_data = ['HeartDisease', 'Smoking', 'AlcoholDrinking', 'Stroke', 
                    'DiffWalking', 'Sex', 'AgeCategory', 'Race', 'Diabetic', 'PhysicalActivity', 'GenHealth', 
                    'Asthma', 'KidneyDisease', 'SkinCancer']

for i in categorical_data:
    print(df[i].unique())
    df[i] = label_encoder.fit_transform(df[i])
    print(df[i].unique())
```

```python
df
```

```python
#drop = hapus kolom
x = np.asarray(df.drop(["HeartDisease", "BMI", "PhysicalHealth", "MentalHealth", "SleepTime"], axis=1))
y = np.asarray(df["HeartDisease"])
```

```python
print(x, y)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

model = DecisionTreeClassifier(criterion = "entropy")
```

```python
model.fit(x, y)
```

```python
plt.subplots(figsize = (12, 10))
plot_tree(model, fontsize = 8)
plt.show()
```
