```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bank.csv", delimiter = ";")
df
```

```python
df.info()
```

```python
#encoding, convert to numierical data

df.columns
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'day', 'month', 'poutcome', 'y']

for col in columns : 
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
#separate data target

x = df.drop('y', axis = 1)
y = df['y']
```

```python
#split data

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
#panggil model KNN

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(x_train, y_train)
```

```python
predicted = model.predict(x_test) #var predicted bakal nampung data hasilnya
```

```python
#Evalusai Model
#ConfusionMatrix berfungsi menghitung jumlah data yang error dan berhasil dari data test dan aktual

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print(confusion_matrix(y_test, predicted))

#benar 9573, 339
#salah 986, 405. Semakin kecil error maka semakin bagus model ML nya
```

```python
print(classification_report(y_test, predicted))
```

```python
print(accuracy_score(y_test, predicted))
```
